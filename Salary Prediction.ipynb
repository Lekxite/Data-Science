```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Survey/survey_results_public.csv")
df.head()
```

### **Let's keep few columns that we need**

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedComp"]]
df = df.rename({"ConvertedComp": "Salary"}, axis = 1)
df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

### **We want to clean the rows where "Employment" is full time**

```python
df = df[df["Employment"] == "Employed full-time"]
df = df.drop("Employment", axis = 1)
df.info()
```

### **Let's clean the country's column**

```python
df["Country"].value_counts()
```

```python
def shorten_categories(categories, cutoff):
  categorical_map = {}
  for i in range(len(categories)):
    if categories.values[i] >= cutoff:
      categorical_map[categories.index[i]] = categories.index[i]
    else:
      categorical_map[categories.index[i]] = "Other"
  return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df["Country"] = df["Country"].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (12, 7))
df.boxplot("Salary", "Country", ax = ax)
plt.suptitle("Salary (US$) vs Country")
plt.title(" ")
plt.ylabel("Country")
plt.xticks(rotation = 90)
plt.show()
```

### **The most datapoint are within the median range of the boxplot.**
### **we want to keep only the data where we have the most information**

```python
df = df[df["Salary"] <= 250000]
df = df[df["Salary"] >= 10000]
df = df[df["Country"] != "Other"]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (12, 7))
df.boxplot("Salary", "Country", ax = ax)
plt.suptitle("Salary (US$) vs Country")
plt.title(" ")
plt.ylabel("Country")
plt.xticks(rotation = 90)
plt.show()
```

### **Now, let's clean the "YearsCodePro" column.**

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(x):
  if x == "More than 50 years":
    return 50
  if x == "Less than 1 year":
    return 0.5
  return float(x)
df["YearsCodePro"] = df["YearsCodePro"].apply(clean_experience)
```

### **Let's also clean the "EdLevel"**

```python
df["EdLevel"].unique()
```

```python
def clean_education(x):
  if "Bachelorâ€™s degree" in x:
    return "Bachelor's degree"
  elif "Masterâ€™s degree" in x:
    return "Masterâ€™s degree"
  elif "Professional degree" in x or "Other doctoral" in x:
    return "Post grad"
  return "Less tha a Bachelor's"

df["EdLevel"] = df["EdLevel"].apply(clean_education)
```

### **Now let's inspect unique values of "EdLevel"**

```python
df["EdLevel"].unique()
```

### **Now we wan to convert the columns that consist of string values**
### **(EdLevel and Country) to numbers that our model can understand**

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df["EdLevel"] = le_education.fit_transform(df["EdLevel"])
df["EdLevel"].unique()
```

```python
le_country = LabelEncoder()
df["Country"] = le_education.fit_transform(df["Country"])
df["Country"].unique()
```

### **Now, let's start training our model. We need to split our data**
### **into features (x) and label (y)**

```python
x = df.drop("Salary", axis = 1)
y = df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x, y.values)
```

### **To predict new values:**

```python
y_pred = linear_reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_pred))
```

### **Actual Label (y) and New Predicted Label (y_pred)**

```python
error
```

### **Let's try another regressor model (DecisionTreeRegressor)**

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state = 0)
dec_tree_reg.fit(x, y.values)
```

### **To predict new values:**

```python
y_pred = dec_tree_reg.predict(x)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

### **Let's try another model**

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state = 0)
random_forest_reg.fit(x, y.values)
```

### **To predict new values:**

```python
y_pred = random_forest_reg.predict(x)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

### **In oder to find the best model with the best parameter, we use**
### **"GridSearchCV" method which stands for Grid Search with**
### **cross-validation**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

max_depth = [None, 2, 4, 6, 8, 10, 12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state = 0)
gs = GridSearchCV(regressor, parameters, scoring = "neg_mean_squared_error")
gs.fit(x, y.values)
```

### **The best estimator is DecisionTreeRegressor**

```python
regressor = gs.best_estimator_

regressor.fit(x, y.values)
y_pred = regressor.predict(x)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
